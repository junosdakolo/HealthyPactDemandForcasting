## Aggregate Monthly Demand

```python
# Aggregate monthly data
df['Month'] = df['Date'].dt.to_period('M').dt.to_timestamp()
monthly_data = df.groupby('Month').agg({
    'Units_Sold': 'sum',
    'CPI': 'mean',
    'Campaign_Flag': 'sum',
    'Holiday_Flag': 'sum'
}).reset_index()
monthly_data.columns = ['ds', 'y', 'CPI', 'Campaigns', 'Holidays']
monthly_data.head()
```

## Build Prophet Model with Regressors

```python
# Initialize Prophet and add regressors
model = Prophet()
model.add_regressor('CPI')
model.add_regressor('Campaigns')
model.add_regressor('Holidays')

# Fit the model
model.fit(monthly_data)
```

## Forecast Future Demand

```python
# Create future dataframe
future = model.make_future_dataframe(periods=6, freq='M')

# Add future values of regressors 
recent = monthly_data.iloc[-6:].mean()
future['CPI'] = recent['CPI']
future['Campaigns'] = recent['Campaigns']
future['Holidays'] = recent['Holidays']

# Predict future demand
forecast = model.predict(future)

# Plot forecast
fig = plot_plotly(model, forecast)
fig.show()
```

## View Forecast Components

```python
# Plot trend and seasonal components
model.plot_components(forecast)
```

# Forecasting Product Demand Using XGBoost (with External Factors)

```python
# Aggregate monthly data
df['Month'] = df['Date'].dt.to_period('M').dt.to_timestamp()
monthly_data = df.groupby('Month').agg({
    'Units_Sold': 'sum',
    'CPI': 'mean',
    'Campaign_Flag': 'sum',
    'Holiday_Flag': 'sum'
}).reset_index()

monthly_data.columns = ['Date', 'Units_Sold', 'CPI', 'Campaigns', 'Holidays']
```

## Prepare Features and Train/Test Split

```python
# Create lag features to mimic temporal effect
monthly_data['Lag_1'] = monthly_data['Units_Sold'].shift(1)
monthly_data = monthly_data.dropna()

# Define features and target
X = monthly_data[['CPI', 'Campaigns', 'Holidays', 'Lag_1']]
y = monthly_data['Units_Sold']

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.25)
```

## Train XGBoost Model

```python
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1)
xgb_model.fit(X_train, y_train)

# Predict
y_pred = xgb_model.predict(X_test)

# Evaluation
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f"XGBoost MAE: {mae:.2f}")
print(f"XGBoost RMSE: {rmse:.2f}")
```

## Actual vs Predicted Plot

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(y_test.index, y_test.values, label='Actual')
plt.plot(y_test.index, y_pred, label='Predicted', linestyle='--')
plt.title('XGBoost: Actual vs Predicted Units Sold')
plt.xlabel('Time Index')
plt.ylabel('Units Sold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot feature importance
plt.figure(figsize=(8, 5))
sns.barplot(x=xgb_model.feature_importances_, y=X.columns)
plt.title('XGBoost Feature Importance')
plt.tight_layout()
plt.show()
```

```python
# Create future dataframe
future = model.make_future_dataframe(periods=6, freq='M')
recent = monthly_data[['CPI', 'Campaigns', 'Holidays']].iloc[-6:].mean()

# Add future values
future['CPI'] = recent['CPI']
future['Campaigns'] = recent['Campaigns']
future['Holidays'] = recent['Holidays']

# Forecast
forecast = model.predict(future)
```
